## 1. Kde jsme teď? Současný stav AI

### Co AI dokáže dnes:
- **Rozpoznávání obrazu** - lepší než člověk
- **Zpracování jazyka** - téměř lidská úroveň
- **Hraní her** - překonává mistry
- **Předpovídání** - přesné predikce v mnoha oblastech
- **Generování obsahu** - text, obrázky, hudba, video

### Co AI zatím neumí:
- **Obecná inteligence** - řešit libovolný problém
- **Skutečné porozumění** - pouze simuluje pochopení
- **Kreativita** - originální myšlenky bez vstupních dat
- **Vědomí** - sebeuvědomění a subjektivní zkušenost

```python
# Vizualizace současných schopností AI
import matplotlib.pyplot as plt
import numpy as np

# Oblasti a úroveň schopností AI vs člověk
oblasti = ['Rozpoznávání\nobrazů', 'Hraní her', 'Překlad', 'Diagnostika\nnemocí', 
           'Kreativita', 'Empatie', 'Zdravý\nrozum', 'Vědomí']
ai_schopnosti = [110, 120, 95, 90, 60, 20, 30, 0]
lidske_schopnosti = [100, 100, 100, 100, 100, 100, 100, 100]

x = np.arange(len(oblasti))
width = 0.35

fig, ax = plt.subplots(figsize=(12, 6))
bars1 = ax.bar(x - width/2, ai_schopnosti, width, label='AI', color='#2196F3')
bars2 = ax.bar(x + width/2, lidske_schopnosti, width, label='Člověk', color='#4CAF50')

# Přidání hodnot na sloupce
for bars in [bars1, bars2]:
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{int(height)}%',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),
                    textcoords="offset points",
                    ha='center', va='bottom', fontsize=9)

ax.set_ylabel('Úroveň schopností (%)', fontsize=12)
ax.set_title('Současné schopnosti AI ve srovnání s člověkem', fontsize=14, fontweight='bold')
ax.set_xticks(x)
ax.set_xticklabels(oblasti)
ax.legend()
ax.grid(True, axis='y', alpha=0.3)
ax.set_ylim(0, 130)

# Přidání čáry lidské úrovně
ax.axhline(y=100, color='red', linestyle='--', alpha=0.5, linewidth=1)

plt.tight_layout()
plt.show()

print("AI již překonává člověka v některých úzkých oblastech,")
print("ale v obecné inteligenci a vědomí za námi výrazně zaostává.")
```

## 2. Tři hlavní scénáře budoucnosti AI

### Scénář 1: Postupný vývoj (Nejpravděpodobnější)

### Scénář 2: Technologická singularita

### Scénář 3: AI plató
- Narazíme na fundamentální limity
- AI zůstane na současné úrovni
- Pouze postupná vylepšení
- Možné už nyní

```python
# Interaktivní časová osa budoucích milníků AI
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from datetime import datetime

# Predikce budoucích milníků
milniky = [
    {"rok": 2025, "udalost": "AI asistenti v každém telefonu", "pravdepodobnost": 95},
    {"rok": 2027, "udalost": "Autonomní auta v běžném provozu", "pravdepodobnost": 80},
    {"rok": 2030, "udalost": "AI lékaři pro základní diagnostiku", "pravdepodobnost": 85},
    {"rok": 2035, "udalost": "AI učitelé pro personalizovanou výuku", "pravdepodobnost": 75},
    {"rok": 2040, "udalost": "Plně automatizované továrny", "pravdepodobnost": 70},
    {"rok": 2045, "udalost": "Obecná AI (AGI)?", "pravdepodobnost": 50},
    {"rok": 2050, "udalost": "AI vědecké objevy", "pravdepodobnost": 60},
]

# Vytvoření grafu
fig, ax = plt.subplots(figsize=(14, 8))

# Nastavení os
current_year = 2024
years = [m["rok"] for m in milniky]
events = [m["udalost"] for m in milniky]
probs = [m["pravdepodobnost"] for m in milniky]

# Barevná škála podle pravděpodobnosti
colors = plt.cm.RdYlGn([p/100 for p in probs])

# Vykreslení časové osy
for i, (year, event, prob, color) in enumerate(zip(years, events, probs, colors)):
    y_pos = i * 0.8
    
    # Kruh pro milník
    circle = plt.Circle((year, y_pos), 0.3, color=color, ec='black', linewidth=2)
    ax.add_patch(circle)
    
    # Popisek události
    ax.text(year + 0.5, y_pos, f"{event}\n({prob}% pravděpodobnost)", 
            va='center', fontsize=10, 
            bbox=dict(boxstyle="round,pad=0.3", facecolor='white', alpha=0.8))
    
    # Spojovací čára
    ax.plot([year, year + 0.5], [y_pos, y_pos], 'k-', alpha=0.3)

# Hlavní časová osa
ax.arrow(current_year-1, -1, 2050-current_year+2, 0, 
         head_width=0.3, head_length=0.5, fc='black', ec='black')

# Současnost
ax.axvline(x=current_year, color='red', linestyle='--', alpha=0.5)
ax.text(current_year, -1.5, 'Současnost', ha='center', fontsize=12, color='red')

# Nastavení grafu
ax.set_xlim(current_year-2, 2052)
ax.set_ylim(-2, len(milniky) * 0.8)
ax.set_xlabel('Rok', fontsize=12)
ax.set_title('Předpokládané milníky budoucnosti AI', fontsize=16, fontweight='bold')
ax.grid(True, axis='x', alpha=0.3)
ax.set_yticks([])

# Legenda pro pravděpodobnost
from matplotlib.patches import Rectangle
legend_elements = [
    Rectangle((0, 0), 1, 1, facecolor=plt.cm.RdYlGn(0.9), label='Vysoká pravděpodobnost (>80%)'),
    Rectangle((0, 0), 1, 1, facecolor=plt.cm.RdYlGn(0.5), label='Střední pravděpodobnost (50-80%)'),
    Rectangle((0, 0), 1, 1, facecolor=plt.cm.RdYlGn(0.2), label='Nízká pravděpodobnost (<50%)')
]
ax.legend(handles=legend_elements, loc='upper left')

plt.tight_layout()
plt.show()
```

   - Podpora nezaměstnaných
   - Rekvalifikační programy
   - Základní příjem?

```python
# Interaktivní plán osobního rozvoje pro budoucnost s AI
class OsobniRozvojAI:
    def __init__(self):
        self.dovednosti = {
            "Technické": {
                "Základy programování": 0,
                "Práce s daty": 0,
                "Pochopení AI": 0,
                "Digitální nástroje": 0
            },
            "Měkké": {
                "Kreativita": 0,
                "Kritické myšlení": 0,
                "Komunikace": 0,
                "Týmová spolupráce": 0,
                "Emoční inteligence": 0
            },
            "Budoucnost": {
                "Adaptabilita": 0,
                "Celoživotní učení": 0,
                "Etické uvažování": 0,
                "Mezioborové znalosti": 0
            }
        }
    
    def ohodnot_dovednost(self, kategorie, dovednost, uroven):
        if kategorie in self.dovednosti and dovednost in self.dovednosti[kategorie]:
            self.dovednosti[kategorie][dovednost] = uroven
    
    def vykresli_profil(self):
        fig, axes = plt.subplots(1, 3, figsize=(18, 6), subplot_kw=dict(projection='polar'))
        
        for ax, (kategorie, dovednosti) in zip(axes, self.dovednosti.items()):
            # Příprava dat
            labels = list(dovednosti.keys())
            values = list(dovednosti.values())
            
            # Počet proměnných
            num_vars = len(labels)
            
            # Výpočet úhlů
            angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
            values += values[:1]
            angles += angles[:1]
            
            # Vytvoření grafu
            ax.plot(angles, values, 'o-', linewidth=2)
            ax.fill(angles, values, alpha=0.25)
            ax.set_xticks(angles[:-1])
            ax.set_xticklabels(labels, size=8)
            ax.set_ylim(0, 10)
            ax.set_title(f'{kategorie} dovednosti', size=12, fontweight='bold', pad=20)
            ax.grid(True)
            
            # Přidání hodnot
            for angle, value, label in zip(angles[:-1], values[:-1], labels):
                ax.text(angle, value + 0.5, str(value), ha='center', va='center', size=8)
        
        plt.suptitle('Můj profil připravenosti na budoucnost s AI', fontsize=16, fontweight='bold')
        plt.tight_layout()
        plt.show()
    
    def doporuceni(self):
        print("\n=== PERSONALIZOVANÁ DOPORUČENÍ ===")
        
        # Najít nejslabší dovednosti
        slabe_dovednosti = []
        for kategorie, dovednosti in self.dovednosti.items():
            for dovednost, uroven in dovednosti.items():
                if uroven < 5:
                    slabe_dovednosti.append((kategorie, dovednost, uroven))
        
        slabe_dovednosti.sort(key=lambda x: x[2])
        
        print("\nDovednosti k rozvoji:")
        for i, (kat, dov, ur) in enumerate(slabe_dovednosti[:3]):
            print(f"{i+1}. {dov} ({kat}) - současná úroveň: {ur}/10")
        
        print("\nDoporučené kroky:")
        print("1. Zapište se na online kurz AI základů")
        print("2. Procvičujte kritické myšlení řešením problémů")
        print("3. Rozvíjejte kreativitu prostřednictvím projektů")
        print("4. Naučte se základy programování (Python)")
        print("5. Sledujte trendy a novinky v AI")

# Ukázka použití
muj_profil = OsobniRozvojAI()

# Simulace hodnocení dovedností (0-10)
# V reálné aplikaci by uživatel vyplnil dotazník
hodnoceni = {
    "Technické": {"Základy programování": 3, "Práce s daty": 4, "Pochopení AI": 6, "Digitální nástroje": 7},
    "Měkké": {"Kreativita": 7, "Kritické myšlení": 6, "Komunikace": 8, "Týmová spolupráce": 7, "Emoční inteligence": 6},
    "Budoucnost": {"Adaptabilita": 5, "Celoživotní učení": 6, "Etické uvažování": 7, "Mezioborové znalosti": 4}
}

for kategorie, dovednosti in hodnoceni.items():
    for dovednost, uroven in dovednosti.items():
        muj_profil.ohodnot_dovednost(kategorie, dovednost, uroven)

muj_profil.vykresli_profil()
muj_profil.doporuceni()
```
