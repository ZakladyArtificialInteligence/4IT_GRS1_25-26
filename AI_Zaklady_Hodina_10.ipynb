## 2. Interaktivn√≠ z√°vƒõreƒçn√Ω kv√≠z

````python
# Komplexn√≠ kv√≠zov√Ω syst√©m s neuronovou s√≠t√≠
import torch
import torch.nn as nn
import gradio as gr
import json
from datetime import datetime

class QuizNeuralNetwork(nn.Module):
    """Neuronov√° s√≠≈• pro adaptivn√≠ kv√≠z"""
    def __init__(self, num_topics=10):
        super().__init__()
        self.fc1 = nn.Linear(num_topics, 32)
        self.fc2 = nn.Linear(32, 16)
        self.fc3 = nn.Linear(16, num_topics)
        self.dropout = nn.Dropout(0.2)
        
    def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = self.dropout(x)
        x = torch.relu(self.fc2(x))
        x = torch.softmax(self.fc3(x), dim=1)
        return x

class AdaptiveQuizSystem:
    def __init__(self):
        self.nn = QuizNeuralNetwork()
        self.user_performance = torch.zeros(10)  # Performance pro ka≈ædou hodinu
        self.questions_bank = self._create_questions_bank()
        self.current_question = None
        self.score = 0
        self.total_questions = 0
        self.history = []
        
    def _create_questions_bank(self):
        """Vytvo≈ôen√≠ banky ot√°zek pro v≈°ech 10 hodin"""
        return {
            1: [  # Hodina 1: Co je AI?
                {
                    "question": "Co znamen√° zkratka AI?",
                    "options": [
                        "Artificial Intelligence",
                        "Automated Interface", 
                        "Advanced Internet",
                        "Analytical Instrument"
                    ],
                    "correct": 0,
                    "explanation": "AI znamen√° Artificial Intelligence (Umƒõl√° inteligence)"
                },
                {
                    "question": "Kter√Ω z n√°sleduj√≠c√≠ch NEN√ç typ strojov√©ho uƒçen√≠?",
                    "options": [
                        "Uƒçen√≠ s uƒçitelem",
                        "Uƒçen√≠ bez uƒçitele",
                        "Posilovan√© uƒçen√≠",
                        "Pasivn√≠ uƒçen√≠"
                    ],
                    "correct": 3,
                    "explanation": "T≈ôi hlavn√≠ typy jsou: s uƒçitelem, bez uƒçitele a posilovan√© uƒçen√≠"
                }
            ],
            2: [  # Hodina 2: Historie AI
                {
                    "question": "Kdo navrhl slavn√Ω test inteligence stroj≈Ø?",
                    "options": [
                        "Alan Turing",
                        "John McCarthy",
                        "Marvin Minsky",
                        "Geoffrey Hinton"
                    ],
                    "correct": 0,
                    "explanation": "Alan Turing navrhl Turing≈Øv test v roce 1950"
                },
                {
                    "question": "Ve kter√©m roce probƒõhla konference v Dartmouth, kter√° zalo≈æila obor AI?",
                    "options": ["1950", "1956", "1960", "1969"],
                    "correct": 1,
                    "explanation": "Dartmouthsk√° konference v roce 1956 je pova≈æov√°na za zrod AI"
                }
            ],
            3: [  # Hodina 3: Budoucnost AI
                {
                    "question": "Co je technologick√° singularita?",
                    "options": [
                        "Bod, kdy AI p≈ôekon√° lidskou inteligenci a zaƒçne se sama vylep≈°ovat",
                        "Prvn√≠ funkƒçn√≠ kvantov√Ω poƒç√≠taƒç",
                        "Konec Moorova z√°kona",
                        "Sjednocen√≠ v≈°ech AI syst√©m≈Ø"
                    ],
                    "correct": 0,
                    "explanation": "Singularita je hypotetick√Ω bod exponenci√°ln√≠ho r≈Østu AI"
                }
            ],
            4: [  # Hodina 4: P≈ô√≠buzn√© obory
                {
                    "question": "Kter√Ω obor poskytuje matematick√Ω z√°klad pro AI?",
                    "options": [
                        "Geologie",
                        "Line√°rn√≠ algebra a statistika",
                        "Organick√° chemie",
                        "Klasick√° mechanika"
                    ],
                    "correct": 1,
                    "explanation": "Line√°rn√≠ algebra a statistika jsou kl√≠ƒçov√© pro ML algoritmy"
                }
            ],
            5: [  # Hodina 5: ƒålovƒõk vs AI
                {
                    "question": "V ƒçem je lidsk√° inteligence st√°le lep≈°√≠ ne≈æ AI?",
                    "options": [
                        "Rychlost v√Ωpoƒçt≈Ø",
                        "Pamƒõ≈• pro fakta",
                        "Kreativita a empatie",
                        "P≈ôesnost v rutinn√≠ch √∫loh√°ch"
                    ],
                    "correct": 2,
                    "explanation": "Lid√© exceluj√≠ v kreativitƒõ, empatii a kontextov√©m porozumƒõn√≠"
                }
            ],
            6: [  # Hodina 6: Etika AI
                {
                    "question": "Co je algoritmick√Ω bias?",
                    "options": [
                        "Chyba v k√≥du",
                        "P≈ôedsudky v datech vedouc√≠ k diskriminaci",
                        "Pomal√Ω algoritmus",
                        "P≈ô√≠li≈° slo≈æit√Ω model"
                    ],
                    "correct": 1,
                    "explanation": "Bias vznik√°, kdy≈æ AI reprodukuje p≈ôedsudky z tr√©novac√≠ch dat"
                }
            ],
            7: [  # Hodina 7: AI v praxi
                {
                    "question": "Kter√° technologie umo≈æ≈àuje Face ID v telefonech?",
                    "options": [
                        "Bluetooth",
                        "NFC",
                        "Neuronov√© s√≠tƒõ pro rozpozn√°v√°n√≠ obliƒçeje",
                        "GPS"
                    ],
                    "correct": 2,
                    "explanation": "Face ID vyu≈æ√≠v√° hlubok√© neuronov√© s√≠tƒõ pro 3D mapov√°n√≠ obliƒçeje"
                }
            ],
            8: [  # Hodina 8: AI ve hr√°ch
                {
                    "question": "Jak√Ω algoritmus pou≈æil Deep Blue pro pora≈æen√≠ Kasparova?",
                    "options": [
                        "Random Forest",
                        "Minimax s alpha-beta pruning",
                        "Genetick√Ω algoritmus",
                        "K-means clustering"
                    ],
                    "correct": 1,
                    "explanation": "Deep Blue pou≈æ√≠val vylep≈°en√Ω minimax algoritmus pro ≈°achy"
                }
            ],
            9: [  # Hodina 9: AI projekt
                {
                    "question": "Co je RAG v kontextu AI?",
                    "options": [
                        "Random Algorithm Generator",
                        "Retrieval-Augmented Generation",
                        "Recursive AI Gateway",
                        "Real-time Analytics Graph"
                    ],
                    "correct": 1,
                    "explanation": "RAG kombinuje vyhled√°v√°n√≠ informac√≠ s generov√°n√≠m textu"
                }
            ],
            10: [  # Hodina 10: Shrnut√≠
                {
                    "question": "Jak√Ω je hlavn√≠ c√≠l kurzu Z√°klady AI?",
                    "options": [
                        "Nauƒçit se programovat neuronov√© s√≠tƒõ",
                        "Pochopit z√°kladn√≠ koncepty a aplikace AI",
                        "Vytvo≈ôit vlastn√≠ AI startup",
                        "Nahradit lidskou pr√°ci"
                    ],
                    "correct": 1,
                    "explanation": "Kurz poskytuje komplexn√≠ √∫vod do svƒõta umƒõl√© inteligence"
                }
            ]
        }
    
    def get_next_question(self):
        """V√Ωbƒõr dal≈°√≠ ot√°zky pomoc√≠ neuronov√© s√≠tƒõ"""
        # Predikce, ze kter√© hodiny vybrat ot√°zku
        with torch.no_grad():
            topic_probs = self.nn(self.user_performance.unsqueeze(0))
        
        # V√Ωbƒõr t√©matu s v√°≈æenou pravdƒõpodobnost√≠
        topic_idx = torch.multinomial(topic_probs, 1).item() + 1
        
        # N√°hodn√Ω v√Ωbƒõr ot√°zky z t√©matu
        if topic_idx in self.questions_bank and self.questions_bank[topic_idx]:
            question_idx = np.random.randint(0, len(self.questions_bank[topic_idx]))
            self.current_question = self.questions_bank[topic_idx][question_idx]
            self.current_question['topic'] = topic_idx
            return self.current_question
        else:
            # Fallback na n√°hodn√© t√©ma
            return self.get_random_question()
    
    def get_random_question(self):
        """N√°hodn√Ω v√Ωbƒõr ot√°zky"""
        topic = np.random.randint(1, 11)
        if self.questions_bank[topic]:
            question_idx = np.random.randint(0, len(self.questions_bank[topic]))
            self.current_question = self.questions_bank[topic][question_idx]
            self.current_question['topic'] = topic
            return self.current_question
    
    def check_answer(self, answer_idx):
        """Kontrola odpovƒõdi a aktualizace performance"""
        if self.current_question is None:
            return False, "Nejprve z√≠skejte ot√°zku"
        
        is_correct = answer_idx == self.current_question['correct']
        self.total_questions += 1
        
        if is_correct:
            self.score += 1
            # Zv√Ω≈°en√≠ performance pro dan√© t√©ma
            self.user_performance[self.current_question['topic'] - 1] += 0.1
        else:
            # Sn√≠≈æen√≠ performance pro dan√© t√©ma
            self.user_performance[self.current_question['topic'] - 1] -= 0.1
        
        # Omezen√≠ hodnot
        self.user_performance = torch.clamp(self.user_performance, -1, 1)
        
        # Ulo≈æen√≠ do historie
        self.history.append({
            'question': self.current_question['question'],
            'correct': is_correct,
            'topic': self.current_question['topic']
        })
        
        return is_correct, self.current_question['explanation']
    
    def get_statistics(self):
        """Z√≠sk√°n√≠ statistik kv√≠zu"""
        if self.total_questions == 0:
            return "Zat√≠m jste nezodpovƒõdƒõli ≈æ√°dnou ot√°zku."
        
        accuracy = (self.score / self.total_questions) * 100
        
        # Statistiky podle t√©mat
        topic_stats = {i: {'correct': 0, 'total': 0} for i in range(1, 11)}
        
        for item in self.history:
            topic = item['topic']
            topic_stats[topic]['total'] += 1
            if item['correct']:
                topic_stats[topic]['correct'] += 1
        
        stats_text = f"""üìä **Va≈°e statistiky**
        
**Celkov√© sk√≥re:** {self.score}/{self.total_questions} ({accuracy:.1f}%)

**V√Ωkon podle t√©mat:**
"""
        
        for topic_id, stats in topic_stats.items():
            if stats['total'] > 0:
                topic_accuracy = (stats['correct'] / stats['total']) * 100
                stats_text += f"Hodina {topic_id}: {stats['correct']}/{stats['total']} ({topic_accuracy:.0f}%)\n"
        
        # Doporuƒçen√≠
        weak_topics = [i+1 for i, perf in enumerate(self.user_performance) if perf < -0.3]
        if weak_topics:
            stats_text += f"\nüí° **Doporuƒçen√≠:** Zopakujte si hodiny: {', '.join(map(str, weak_topics))}"
        
        return stats_text

# Vytvo≈ôen√≠ kv√≠zov√©ho syst√©mu
quiz_system = AdaptiveQuizSystem()

# Gradio interface pro kv√≠z
def quiz_interface():
    with gr.Blocks(title="AI Kv√≠z - Z√°vƒõreƒçn√Ω test") as demo:
        gr.Markdown("""
        # üéØ Z√°vƒõreƒçn√Ω kv√≠z - Z√°klady AI
        
        Otestujte sv√© znalosti z prvn√≠ch 10 hodin kurzu!
        Kv√≠z se adaptuje podle va≈°ich odpovƒõd√≠.
        """)
        
        with gr.Row():
            with gr.Column(scale=2):
                question_display = gr.Textbox(
                    label="Ot√°zka",
                    interactive=False,
                    lines=3
                )
                
                options_radio = gr.Radio(
                    label="Vyberte odpovƒõƒè",
                    choices=[],
                    type="index"
                )
                
                with gr.Row():
                    next_btn = gr.Button("üìù Dal≈°√≠ ot√°zka", variant="primary")
                    submit_btn = gr.Button("‚úÖ Odpovƒõdƒõt", variant="secondary")
                
                feedback = gr.Textbox(
                    label="Zpƒõtn√° vazba",
                    interactive=False,
                    lines=2
                )
            
            with gr.Column(scale=1):
                stats_display = gr.Textbox(
                    label="Statistiky",
                    interactive=False,
                    lines=15
                )
                
                refresh_stats_btn = gr.Button("üîÑ Aktualizovat statistiky")
        
        def get_question():
            """Z√≠sk√°n√≠ nov√© ot√°zky"""
            q = quiz_system.get_next_question()
            return (
                f"Hodina {q['topic']}: {q['question']}",
                gr.update(choices=q['options'], value=None),
                ""
            )
        
        def submit_answer(answer_idx):
            """Odesl√°n√≠ odpovƒõdi"""
            if answer_idx is None:
                return "Pros√≠m vyberte odpovƒõƒè"
            
            is_correct, explanation = quiz_system.check_answer(answer_idx)
            
            if is_correct:
                return f"‚úÖ Spr√°vnƒõ! {explanation}"
            else:
                return f"‚ùå ≈†patnƒõ. {explanation}"
        
        # Propojen√≠ funkc√≠
        next_btn.click(
            fn=get_question,
            outputs=[question_display, options_radio, feedback]
        )
        
        submit_btn.click(
            fn=submit_answer,
            inputs=options_radio,
            outputs=feedback
        )
        
        refresh_stats_btn.click(
            fn=lambda: quiz_system.get_statistics(),
            outputs=stats_display
        )
        
        # Naƒçten√≠ prvn√≠ ot√°zky p≈ôi spu≈°tƒõn√≠
        demo.load(
            fn=get_question,
            outputs=[question_display, options_radio, feedback]
        )
    
    return demo

# Spu≈°tƒõn√≠ kv√≠zu
print("üöÄ Spou≈°t√≠m interaktivn√≠ kv√≠z...")
quiz_demo = quiz_interface()
quiz_demo.launch(share=True)
````

## 3. Praktick√° uk√°zka v≈°ech nauƒçen√Ωch technik

````python
# Showcase v≈°ech technik v jedn√© aplikaci
import torch
import torch.nn as nn
import numpy as np
from transformers import pipeline
import cv2
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

class AITechniquesShowcase:
    def __init__(self):
        print("üé™ Inicializuji showcase AI technik...")
        self.techniques = {
            "neural_networks": "Neuronov√© s√≠tƒõ",
            "computer_vision": "Poƒç√≠taƒçov√© vidƒõn√≠",
            "nlp": "Zpracov√°n√≠ p≈ôirozen√©ho jazyka",
            "reinforcement_learning": "Posilovan√© uƒçen√≠",
            "clustering": "Shlukov√°n√≠",
            "decision_trees": "Rozhodovac√≠ stromy"
        }
        
    def demo_neural_network(self):
        """Uk√°zka jednoduch√© neuronov√© s√≠tƒõ"""
        print("\nüß† DEMO: Neuronov√° s√≠≈• pro XOR probl√©m")
        
        # Definice s√≠tƒõ
        class XORNet(nn.Module):
            def __init__(self):
                super().__init__()
                self.fc1 = nn.Linear(2, 4)
                self.fc2 = nn.Linear(4, 1)
                
            def forward(self, x):
                x = torch.sigmoid(self.fc1(x))
                x = torch.sigmoid(self.fc2(x))
                return x
        
        # Tr√©novac√≠ data pro XOR
        X = torch.tensor([[0, 0], [0, 1], [1, 0], [1, 1]], dtype=torch.float32)
        y = torch.tensor([[0], [1], [1], [0]], dtype=torch.float32)
        
        # Tr√©nov√°n√≠
        model = XORNet()
        optimizer = torch.optim.Adam(model.parameters(), lr=0.1)
        criterion = nn.BCELoss()
        
        for epoch in range(1000):
            output = model(X)
            loss = criterion(output, y)
            
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
        
        # Test
        with torch.no_grad():
            predictions = model(X)
            print("\nVstupy | Oƒçek√°v√°no | Predikce")
            print("-" * 35)
            for i in range(len(X)):
                print(f"{X[i].numpy()} | {y[i].item():.0f} | {predictions[i].item():.3f}")
        
        return model
    
    def demo_computer_vision(self):
        """Uk√°zka detekce hran"""
        print("\nüëÅÔ∏è DEMO: Poƒç√≠taƒçov√© vidƒõn√≠ - Detekce hran")
        
        # Vytvo≈ôen√≠ syntetick√©ho obr√°zku
        img = np.zeros((200, 200), dtype=np.uint8)
        cv2.rectangle(img, (50, 50), (150, 150), 255, -1)
        cv2.circle(img, (100, 100), 30, 0, -1)
        
        # Detekce hran
        edges = cv2.Canny(img, 50, 150)
        
        # Vizualizace
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
        ax1.imshow(img, cmap='gray')
        ax1.set_title('P≈Øvodn√≠ obr√°zek')
        ax1.axis('off')
        
        ax2.imshow(edges, cmap='gray')
        ax2.set_title('Detekovan√© hrany')
        ax2.axis('off')
        
        plt.tight_layout()
        plt.show()
        
        return edges
    
    def demo_nlp(self):
        """Uk√°zka anal√Ωzy sentimentu"""
        print("\nüí¨ DEMO: NLP - Anal√Ωza sentimentu")
        
        # Jednoduch√Ω sentiment analyzer
        positive_words = ['skvƒõl√Ω', 'v√Ωborn√Ω', 'super', 'kr√°sn√Ω', '√∫≈æasn√Ω']
        negative_words = ['≈°patn√Ω', 'hrozn√Ω', 'zklam√°n√≠', 'nudn√Ω', 'slab√Ω']
        
        texts = [
            "Tento kurz AI je naprosto skvƒõl√Ω!",
            "Bylo to hrozn√© zklam√°n√≠.",
            "Pr≈Ømƒõrn√Ω v√Ωsledek, nic zvl√°≈°tn√≠ho."
        ]
        
        print("\nAnal√Ωza sentimentu:")
        for text in texts:
            positive_score = sum(1 for word in positive_words if word in text.lower())
            negative_score = sum(1 for word in negative_words if word in text.lower())
            
            if positive_score > negative_score:
                sentiment = "üòä Pozitivn√≠"
            elif negative_score > positive_score:
                sentiment = "üòû Negativn√≠"
            else:
                sentiment = "üòê Neutr√°ln√≠"
            
            print(f"Text: '{text}'")
            print(f"Sentiment: {sentiment}\n")
    
    def demo_clustering(self):
        """Uk√°zka shlukov√°n√≠"""
        print("\nüîÆ DEMO: Shlukov√°n√≠ - K-means")
        
        # Generov√°n√≠ dat
        np.random.seed(42)
        cluster1 = np.random.randn(50, 2) + [2, 2]
        cluster2 = np.random.randn(50, 2) + [-2, -2]
        cluster3 = np.random.randn(50, 2) + [2, -2]
        
        X = np.vstack([cluster1, cluster2, cluster3])
        
        # K-means clustering
        kmeans = KMeans(n_clusters=3, random_state=42)
        labels = kmeans.fit_predict(X)
        
        # Vizualizace
        plt.figure(figsize=(8, 6))
        scatter = plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', alpha=0.6)
        centers = kmeans.cluster_centers_
        plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, marker='*', 
                   edgecolor='black', linewidth=2)
        plt.title('K-means Clustering', fontsize=14)
        plt.xlabel('Feature 1')
        plt.ylabel('Feature 2')
        plt.colorbar(scatter)
        plt.grid(True, alpha=0.3)
        plt.show()
        
        return kmeans
    
    def run_all_demos(self):
        """Spu≈°tƒõn√≠ v≈°ech uk√°zek"""
        print("üé≠ SHOWCASE V≈†ECH AI TECHNIK")
        print("=" * 50)
        
        # 1. Neuronov√© s√≠tƒõ
        self.demo_neural_network()
        
        # 2. Poƒç√≠taƒçov√© vidƒõn√≠
        self.demo_computer_vision()
        
        # 3. NLP
        self.demo_nlp()
        
        # 4. Shlukov√°n√≠
        self.demo_clustering()
        
        print("\n‚úÖ V≈°echny techniky √∫spƒõ≈°nƒõ p≈ôedvedeny!")

# Spu≈°tƒõn√≠ showcase
showcase = AITechniquesShowcase()
showcase.run_all_demos()
````
