```python
# Ukázka důležitosti algoritmů pro AI
import time
import random
import matplotlib.pyplot as plt

# Porovnání různých algoritmů vyhledávání
def linearni_vyhledavani(seznam, hledany):
    """Lineární vyhledávání - O(n)"""
    for i, prvek in enumerate(seznam):
        if prvek == hledany:
            return i
    return -1

def binarni_vyhledavani(seznam, hledany):
    """Binární vyhledávání - O(log n)"""
    levy, pravy = 0, len(seznam) - 1
    
    while levy <= pravy:
        stred = (levy + pravy) // 2
        if seznam[stred] == hledany:
            return stred
        elif seznam[stred] < hledany:
            levy = stred + 1
        else:
            pravy = stred - 1
    return -1

# Test výkonu
velikosti = [100, 1000, 10000, 50000, 100000]
cas_linearni = []
cas_binarni = []

for velikost in velikosti:
    # Vytvoření testovacích dat
    data = sorted(random.sample(range(velikost * 10), velikost))
    hledany = random.choice(data)
    
    # Lineární vyhledávání
    start = time.time()
    linearni_vyhledavani(data, hledany)
    cas_linearni.append(time.time() - start)
    
    # Binární vyhledávání
    start = time.time()
    binarni_vyhledavani(data, hledany)
    cas_binarni.append(time.time() - start)

# Vizualizace
plt.figure(figsize=(10, 6))
plt.plot(velikosti, cas_linearni, 'o-', label='Lineární vyhledávání O(n)', linewidth=2)
plt.plot(velikosti, cas_binarni, 's-', label='Binární vyhledávání O(log n)', linewidth=2)
plt.xlabel('Velikost dat', fontsize=12)
plt.ylabel('Čas (sekundy)', fontsize=12)
plt.title('Význam efektivních algoritmů pro AI', fontsize=14, fontweight='bold')
plt.legend()
plt.grid(True, alpha=0.3)
plt.yscale('log')
plt.show()

print("Pro AI pracující s velkými daty je volba správného algoritmu klíčová!")
print(f"Při 100,000 prvcích je binární vyhledávání {cas_linearni[-1]/cas_binarni[-1]:.0f}x rychlejší")
```

```python
# 4.4 Kódování a testování — jednoduchý ML pipeline se scikit-learn
# Cíl: ukázat předzpracování, trénink, testování a jednoduchý test pro ověření výsledku
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import matplotlib.pyplot as plt
import numpy as np

# Načtení dat
iris = load_iris()
X, y = iris.data, iris.target

# Train/test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42, stratify=y)

# Pipeline: škálování + logistická regrese
pipeline = Pipeline([
    ("scaler", StandardScaler()),
    ("clf", LogisticRegression(solver="liblinear", random_state=42, max_iter=200))
])

# Trénink
pipeline.fit(X_train, y_train)

# Predikce
y_pred = pipeline.predict(X_test)

# Vyhodnocení
acc = accuracy_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)
print(f"Velikost trénovací množiny: {len(y_train)}, testovací množiny: {len(y_test)}")
print(f"Accuracy na test setu: {acc:.3f}")
print("\nClassification report:")
print(classification_report(y_test, y_pred, target_names=iris.target_names))

# Jednoduchý test (ukázka automatického ověření studentů)
# Očekáváme rozumnou přesnost na Iris; nastavíme prah 0.7
assert acc >= 0.7, f"Accuracy je příliš nízká: {acc:.3f} (< 0.7)"
print("\n✔ Test: accuracy >= 0.7 — splněno")

# Volitelná křížová validace
cv_scores = cross_val_score(pipeline, X, y, cv=5)
print(f"5-fold cross-val accuracy: {np.mean(cv_scores):.3f} ± {np.std(cv_scores):.3f}")

# Vizualizace confusion matrix
plt.figure(figsize=(5, 4))
plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
plt.title('Confusion matrix')
plt.colorbar()
classes = iris.target_names
plt.xticks(np.arange(len(classes)), classes, rotation=45)
plt.yticks(np.arange(len(classes)), classes)
plt.xlabel('Predicted')
plt.ylabel('True')
for i in range(cm.shape[0]):
    for j in range(cm.shape[1]):
        plt.text(j, i, cm[i, j], horizontalalignment='center', color='white' if cm[i, j] > cm.max()/2 else 'black')
plt.tight_layout()
plt.show()

# Poznámky pro studenty:
# - Předzpracování (StandardScaler) je často důležité pro modely citlivé na měřítko
# - Rozdělení na trénink/test je nutné pro poctivé vyhodnocení
# - Křížová validace dává robustnější odhad výkonu na celém datasetu
```

## 4.2 AI vs. statistika

AI a statistika mají překryv, ale liší se v cílech a metodách:

- Cíle:
  - Statistika: pochopení dat, inference, testování hypotéz, odhad parametrů.
  - AI (strojové učení): predikce a automatizace rozhodování, často optimalizace výkonu na velkých datech.

- Metody:
  - Statistika používá modely zaměřené na interpretovatelnost a odhady nejistoty (např. intervaly spolehlivosti, p-hodnoty).
  - AI často používá komplexní modely (např. neuronové sítě), které dávají lepší predikční výkon, ale méně interpretovatelné výstupy.

- Příklad rozdílu:
  - Statistický přístup: zjistit, zda existuje vztah mezi proměnnými (test hypotézy).
  - AI přístup: vyrobit model, který co nejlépe předpovídá budoucí hodnotu bez nutnosti jednoduché interpretace všech parametrů.

```python
# Ukázka matematiky v AI - gradientní sestup
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Definice jednoduché funkce pro optimalizaci
def funkce(x, y):
    """Funkce s lokálními minimy"""
    return x**2 + y**2 + 3*np.sin(2*x) + 3*np.sin(2*y)

def gradient(x, y):
    """Gradient funkce"""
    dx = 2*x + 6*np.cos(2*x)
    dy = 2*y + 6*np.cos(2*y)
    return dx, dy

# Gradientní sestup
def gradientni_sestup(start_x, start_y, learning_rate=0.1, kroky=50):
    x, y = start_x, start_y
    cesta = [(x, y, funkce(x, y))]
    
    for _ in range(kroky):
        dx, dy = gradient(x, y)
        x -= learning_rate * dx
        y -= learning_rate * dy
        cesta.append((x, y, funkce(x, y)))
    
    return np.array(cesta)

# Vytvoření 3D grafu
fig = plt.figure(figsize=(14, 6))

# Graf 1: 3D povrch funkce
ax1 = fig.add_subplot(121, projection='3d')

x_range = np.linspace(-5, 5, 100)
y_range = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x_range, y_range)
Z = funkce(X, Y)

ax1.plot_surface(X, Y, Z, cmap='viridis', alpha=0.6)

# Vykreslení cest gradientního sestupu z různých bodů
start_points = [(3, 3), (-3, 3), (3, -3), (-3, -3)]
colors = ['red', 'blue', 'green', 'orange']

for start, color in zip(start_points, colors):
    cesta = gradientni_sestup(start[0], start[1])
    ax1.plot(cesta[:, 0], cesta[:, 1], cesta[:, 2], 
             color=color, linewidth=2, marker='o', markersize=3)

ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('f(x,y)')
ax1.set_title('Gradientní sestup v 3D', fontsize=12)

# Graf 2: Kontury a cesty
ax2 = fig.add_subplot(122)

contour = ax2.contour(X, Y, Z, levels=20, cmap='viridis')
ax2.clabel(contour, inline=True, fontsize=8)

for start, color in zip(start_points, colors):
    cesta = gradientni_sestup(start[0], start[1])
    ax2.plot(cesta[:, 0], cesta[:, 1], 
             color=color, linewidth=2, marker='o', markersize=3,
             label=f'Start: {start}')
    ax2.scatter(cesta[0, 0], cesta[0, 1], s=100, color=color, marker='x')
    ax2.scatter(cesta[-1, 0], cesta[-1, 1], s=100, color=color, marker='*')

ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('Cesty gradientního sestupu', fontsize=12)
ax2.legend()
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("Gradientní sestup je základní matematická metoda používaná při trénování neuronových sítí.")
print("x = začátek, * = konec cesty")
```

## 4. Psychologie a kognitivní vědy

### Přínosy psychologie pro AI:

1. **Modely učení**
   - Podmíněné učení
   - Posilované učení

2. **Paměť a pozornost**
   - Krátkodobá vs. dlouhodobá paměť
   - Mechanismy pozornosti

3. **Rozhodování**
   - Heuristiky a zkreslení
   - Emoce v rozhodování

4. **Vnímání**
   - Vizuální zpracování
   - Rozpoznávání vzorů

```python
# Ukázka psychologického konceptu v AI - paměť a zapomínání
import numpy as np
import matplotlib.pyplot as plt

class KognitivniModel:
    def __init__(self):
        self.kratkodoba_pamet = []  # Max 7±2 položek
        self.dlouhodoba_pamet = {}
        self.cas = 0
    
    def zapamatuj(self, informace, dulezitost=1):
        """Uložení informace do paměti"""
        # Millerovo magické číslo 7±2
        if len(self.kratkodoba_pamet) >= 7:
            # Přesun nejstarší položky do dlouhodobé paměti nebo zapomenutí
            nejstarsi = self.kratkodoba_pamet.pop(0)
            if nejstarsi['dulezitost'] > 0.5:
                self.dlouhodoba_pamet[nejstarsi['info']] = {
                    'sila': nejstarsi['dulezitost'],
                    'cas_ulozeni': nejstarsi['cas']
                }
        
        self.kratkodoba_pamet.append({
            'info': informace,
            'dulezitost': dulezitost,
            'cas': self.cas
        })
        self.cas += 1
    
    def vybav_si(self, informace):
        """Vybavení informace z paměti"""
        # Hledání v krátkodobé paměti
        for polozka in self.kratkodoba_pamet:
            if polozka['info'] == informace:
                return True, "krátkodobá paměť"
        
        # Hledání v dlouhodobé paměti s Ebbinghausovou křivkou zapomínání
        if informace in self.dlouhodoba_pamet:
            cas_od_ulozeni = self.cas - self.dlouhodoba_pamet[informace]['cas_ulozeni']
            sila_puvodni = self.dlouhodoba_pamet[informace]['sila']
            
            # Ebbinghausova křivka zapomínání
            sila_aktualni = sila_puvodni * np.exp(-cas_od_ulozeni / 10)
            
            if sila_aktualni > 0.3:  # Práh pro vybavení
                return True, f"dlouhodobá paměť (síla: {sila_aktualni:.2f})"
        
        return False, "zapomenuto"

# Demonstrace modelu paměti
model = KognitivniModel()

# Simulace učení
informace = [
    ("Python je programovací jazyk", 0.9),
    ("AI znamená umělá inteligence", 0.95),
    ("Neurony jsou základem mozku", 0.7),
    ("2+2=4", 0.6),
    ("Hlavní město ČR je Praha", 0.8),
    ("Voda vaří při 100°C", 0.7),
    ("Slunce je hvězda", 0.75),
    ("DNA nese genetickou informaci", 0.85),
    ("Fotosyntéza vytváří kyslík", 0.65)
]

print("=== SIMULACE KOGNITIVNÍHO MODELU PAMĚTI ===")
print("\nUčení informací:")
for info, dulezitost in informace:
    model.zapamatuj(info, dulezitost)
    print(f"✓ Naučeno: {info[:30]}... (důležitost: {dulezitost})")

print(f"\nKrátkodobá paměť obsahuje {len(model.kratkodoba_pamet)} položek")
print(f"Dlouhodobá paměť obsahuje {len(model.dlouhodoba_pamet)} položek")

# Simulace času a vybavování
print("\n=== TEST VYBAVOVÁNÍ PO ČASE ===")

# Posun času
model.cas += 15

test_informace = [
    "Python je programovací jazyk",
    "2+2=4",
    "DNA nese genetickou informaci",
    "Něco co jsme se neučili"
]

for info in test_informace:
    nalezeno, zdroj = model.vybav_si(info)
    if nalezeno:
        print(f"✓ '{info[:30]}...' - nalezeno v {zdroj}")
    else:
        print(f"✗ '{info[:30]}...' - {zdroj}")

# Vizualizace Ebbinghausovy křivky zapomínání
plt.figure(figsize=(10, 6))

cas = np.linspace(0, 50, 100)
sily = [1.0, 0.8, 0.6, 0.4]
labels = ['Velmi důležité (1.0)', 'Důležité (0.8)', 'Středně důležité (0.6)', 'Méně důležité (0.4)']

for sila, label in zip(sily, labels):
    pamet = sila * np.exp(-cas / 10)
    plt.plot(cas, pamet, linewidth=2, label=label)

plt.axhline(y=0.3, color='red', linestyle='--', alpha=0.5, label='Práh vybavení')
plt.xlabel('Čas', fontsize=12)
plt.ylabel('Síla paměti', fontsize=12)
plt.title('Ebbinghausova křivka zapomínání v AI modelu', fontsize=14, fontweight='bold')
plt.legend()
plt.grid(True, alpha=0.3)
plt.ylim(0, 1.1)
plt.show()

print("\nPsychologické principy pomáhají vytvářet efektivnější AI systémy!")
```

## 5. Neurovědy a biologická inspirace

### Jak mozek inspiruje AI:

1. **Neuronové sítě**
   - Struktura neuronů
   - Synapse a učení

2. **Specializované oblasti**
   - Vizuální kortex → CNN
   - Hippocampus → paměťové sítě

3. **Plasticita**
   - Adaptace a učení
   - Reorganizace spojení

```python
# Simulace biologického neuronu vs. umělý neuron
import numpy as np
import matplotlib.pyplot as plt

class BiologickyNeuron:
    def __init__(self):
        self.prah = -55  # mV - práh pro akční potenciál
        self.klidovy_potencial = -70  # mV
        self.potencial = self.klidovy_potencial
        self.refrakcni_doba = 0
        self.historie = []
    
    def prijmi_signal(self, vstup):
        """Simulace příjmu signálu"""
        if self.refrakcni_doba > 0:
            self.refrakcni_doba -= 1
            return 0
        
        # Změna potenciálu
        self.potencial += vstup
        
        # Kontrola prahu
        if self.potencial >= self.prah:
            # Akční potenciál!
            vystup = 1
            self.potencial = self.klidovy_potencial
            self.refrakcni_doba = 3
        else:
            vystup = 0
            # Postupný návrat ke klidovému potenciálu
            self.potencial += (self.klidovy_potencial - self.potencial) * 0.1
        
        self.historie.append(self.potencial)
        return vystup

class UmelyNeuron:
    def __init__(self, pocet_vstupu):
        self.vahy = np.random.randn(pocet_vstupu) * 0.1
        self.bias = 0
    
    def aktivacni_funkce(self, x):
        """Sigmoid aktivační funkce"""
        return 1 / (1 + np.exp(-x))
    
    def vypocet(self, vstupy):
        """Výpočet výstupu neuronu"""
        suma = np.dot(vstupy, self.vahy) + self.bias
        return self.aktivacni_funkce(suma)

# Simulace
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(12, 10))

# 1. Biologický neuron
bio_neuron = BiologickyNeuron()
vstupy = np.random.normal(5, 2, 100)  # Náhodné vstupní signály
vystupy_bio = []

for vstup in vstupy:
    vystup = bio_neuron.prijmi_signal(vstup)
    vystupy_bio.append(vystup)

ax1.plot(bio_neuron.historie, color='blue', linewidth=1)
ax1.axhline(y=bio_neuron.prah, color='red', linestyle='--', label='Práh')
ax1.axhline(y=bio_neuron.klidovy_potencial, color='green', linestyle='--', label='Klidový potenciál')
ax1.set_ylabel('Membránový potenciál (mV)')
ax1.set_title('Biologický neuron - membránový potenciál', fontsize=12)
ax1.legend()
ax1.grid(True, alpha=0.3)

# Výstupy biologického neuronu
ax2.stem(vystupy_bio, basefmt=' ')
ax2.set_ylabel('Akční potenciály')
ax2.set_title('Výstup biologického neuronu (spiky)', fontsize=12)
ax2.set_ylim(-0.1, 1.1)
ax2.grid(True, alpha=0.3)

# 2. Umělý neuron
umely_neuron = UmelyNeuron(3)
x = np.linspace(-5, 5, 100)
y = 1 / (1 + np.exp(-x))  # Sigmoid

ax3.plot(x, y, 'g-', linewidth=2, label='Sigmoid aktivační funkce')
ax3.axhline(y=0.5, color='red', linestyle='--', alpha=0.5)
ax3.axvline(x=0, color='red', linestyle='--', alpha=0.5)
ax3.set_xlabel('Vstup')
ax3.set_ylabel('Výstup')
ax3.set_title('Umělý neuron - aktivační funkce', fontsize=12)
ax3.grid(True, alpha=0.3)
ax3.legend()

plt.tight_layout()
plt.show()

# Porovnání
print("=== POROVNÁNÍ BIOLOGICKÉHO A UMĚLÉHO NEURONU ===")
print("\nBiologický neuron:")
print("- Pracuje s elektrickými impulsy (akční potenciály)")
print("- Má refrakterní dobu (nemůže pálit neustále)")
print("- Komunikuje pomocí diskrétních pulsů")
print("\nUmělý neuron:")
print("- Pracuje s kontinuálními hodnotami")
print("- Okamžitá odpověď bez refrakterní doby")
print("- Výstup je spojitá funkce vstupů")
print("\nPřestože jsou rozdílné, oba implementují podobný princip:")
print("integraci vstupů a prahové rozhodování!")
```

## 6. Lingvistika a zpracování jazyka

### Přínosy lingvistiky:

1. **Syntaxe**
   - Gramatická pravidla
   - Struktura vět

2. **Sémantika**
   - Význam slov
   - Kontextové porozumění

3. **Pragmatika**
   - Použití jazyka v kontextu
   - Komunikační záměr

```python
# Ukázka lingvistické analýzy pro AI
import re
from collections import Counter

class LingvistickyAnalyzator:
    def __init__(self):
        # Česká slovní zásoba
        self.slovni_druhy = {
            'podstatna': ['pes', 'kočka', 'dům', 'strom', 'auto', 'člověk', 'kniha'],
            'slovesa': ['běží', 'spí', 'jí', 'čte', 'píše', 'hraje', 'zpívá'],
            'pridavna': ['velký', 'malý', 'červený', 'modrý', 'rychlý', 'chytrý'],
            'predlozky': ['na', 'v', 'pod', 'nad', 'před', 'za', 'mezi'],
            'spojky': ['a', 'ale', 'nebo', 'protože', 'když', 'že']
        }
        
        # Syntaktická pravidla (zjednodušená)
        self.pravidla = [
            r'(\w+)\s+(běží|spí|jí|čte|píše|hraje|zpívá)',  # Podmět + sloveso
            r'(velký|malý|červený|modrý|rychlý|chytrý)\s+(pes|kočka|dům|strom|auto|člověk)',  # Přídavné + podstatné
        ]
    
    def analyzuj_vetu(self, veta):
        """Základní lingvistická analýza věty"""
        veta = veta.lower()
        slova = veta.split()
        
        analyza = {
            'pocet_slov': len(slova),
            'slovni_druhy': {},
            'syntakticke_vzory': []
        }
        
        # Analýza slovních druhů
        for slovo in slova:
            for druh, seznam in self.slovni_druhy.items():
                if slovo in seznam:
                    if druh not in analyza['slovni_druhy']:
                        analyza['slovni_druhy'][druh] = []
                    analyza['slovni_druhy'][druh].append(slovo)
        
        # Hledání syntaktických vzorů
        for pravidlo in self.pravidla:
            shody = re.findall(pravidlo, veta)
            if shody:
                analyza['syntakticke_vzory'].extend(shody)
        
        return analyza
    
    def semanticka_podobnost(self, slovo1, slovo2):
        """Jednoduchá sémantická podobnost na základě kategorie"""
        for druh, seznam in self.slovni_druhy.items():
            if slovo1 in seznam and slovo2 in seznam:
                return 0.8  # Vysoká podobnost - stejná kategorie
        return 0.2  # Nízká podobnost

# Demonstrace
analyzer = LingvistickyAnalyzator()

# Testovací věty
vety = [
    "Velký pes běží rychle",
    "Malá kočka spí na střeše",
    "Červený auto stojí před domem",
    "Chytrý člověk čte knihu"
]

print("=== LINGVISTICKÁ ANALÝZA PRO AI ===")

for veta in vety:
    print(f"\nVěta: '{veta}'")
    analyza = analyzer.analyzuj_vetu(veta)
    
    print(f"Počet slov: {analyza['pocet_slov']}")
    
    if analyza['slovni_druhy']:
        print("Rozpoznané slovní druhy:")
        for druh, slova in analyza['slovni_druhy'].items():
            print(f"  - {druh}: {', '.join(slova)}")
    
    if analyza['syntakticke_vzory']:
        print("Syntaktické vzory:")
        for vzor in analyza['syntakticke_vzory']:
            print(f"  - {' '.join(vzor) if isinstance(vzor, tuple) else vzor}")

# Sémantická podobnost
print("\n=== SÉMANTICKÁ PODOBNOST ===")
test_pary = [("pes", "kočka"), ("pes", "běží"), ("velký", "malý")]

for slovo1, slovo2 in test_pary:
    podobnost = analyzer.semanticka_podobnost(slovo1, slovo2)
    print(f"{slovo1} <-> {slovo2}: {podobnost:.1%} podobnost")

# Vizualizace frekvenční analýzy
plt.figure(figsize=(10, 6))

# Spojení všech vět
vsechny_slova = ' '.join(vety).lower().split()
frekvence = Counter(vsechny_slova)

slova = list(frekvence.keys())
pocty = list(frekvence.values())

plt.bar(slova, pocty, color='skyblue')
plt.xlabel('Slova', fontsize=12)
plt.ylabel('Frekvence', fontsize=12)
plt.title('Frekvenční analýza slov', fontsize=14, fontweight='bold')
plt.xticks(rotation=45, ha='right')
plt.grid(True, axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

print("\nLingvistika poskytuje AI nástroje pro pochopení struktury a významu jazyka!")
```
